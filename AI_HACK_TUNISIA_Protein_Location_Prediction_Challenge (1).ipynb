Here for the first model (Nucleus), we deleted all other classes

```python
pre_train_Nucleus = train.drop(['Cytoplasm', 'Membrane', 'Cell membrane', 'Extracellular'], axis=1)
pre_train_Nucleus.head()
```

```python
df2 = pd.read_csv('/content/sequence_esm1b_mean_embeddings_df.csv',index_col=0)
```

```python
pre_train_Nucleus.join(df2, on="sequence", how="inner")
train_Nucleus = pre_train_Nucleus.merge(df2, left_on="sequence", right_on="sequence")
train_Nucleus = train_Nucleus.drop(columns=["sequence", "Kingdom", "seq_len"])
train_Nucleus.head()
```

```python
from sklearn.preprocessing import LabelEncoder
def Encoder(df):
          columnsToEncode = list(df.select_dtypes(include=['category','object']))
          le = LabelEncoder()
          for feature in columnsToEncode:
              try:
                  df[feature] = le.fit_transform(df[feature])
              except:
                  print('Error encoding '+feature)
          return df
```

```python
Encoder(train_Nucleus)
```

```python
X_Nucleus = train_Nucleus.drop(['Nucleus'], axis=1)
Y_Nucleus = train_Nucleus['Nucleus']
```

```python
X_Nucleus_n=normalize(X_Nucleus)
```

```python
X_Nucleus_train, X_Nucleus_test, y_Nucleus_train, y_Nucleus_test = train_test_split(X_Nucleus_n, Y_Nucleus, test_size=0.2, random_state=8)
```

### Building the first model to predict Nucleus class (Neural Network model)

```python
model_Nucleus = Sequential()
model_Nucleus.add(Dense(units=1280,  activation='relu', input_shape=(1280,)))
model_Nucleus.add(Dense(units=1, activation='sigmoid'))
model_Nucleus.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history_Nucleus = model_Nucleus.fit(X_Nucleus_train, y_Nucleus_train, epochs=20, validation_data=(X_Nucleus_test, y_Nucleus_test))
```

### Building the second model to predict Nucleus class (XGB model)

```python
data_dmatrix = xgb.DMatrix(data=X_Nucleus_train,label=y_Nucleus_train)
xgc_Nucleus = XGBClassifier(max_depth=10,colsample_bytree=0.8,learning_rate=0.02,n_estimators=500,nthread=-1)
# xgc_Nucleus = xgb.XGBClassifier(objective ='reg:logistic', colsample_bytree = 0.15,
#                           learning_rate = 0.1, 
#                           max_depth = 30, alpha = 12, n_estimators = 700)
xgc_Nucleus.fit(X_Nucleus_train,(y_Nucleus_train),eval_metric="logloss",eval_set=[(X_Nucleus_test, y_Nucleus_test)])
#xgc_Nucleus.fit(X_Nucleus_train,y_Nucleus_train)
```
